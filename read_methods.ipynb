```python
import pandas as pd
import numpy as np
import pickle
import os
```

```python
filename = "load.csv"
try:
    os.chdir(os.path.join(os.getcwd(), 'resources/LoadingDatasets'))
    print(os.getcwd())
except:
    pass
```

```python
cols = None
data = []
```

# method0 - read line by line

```python
with open(filename) as f:
    for line in f.readlines():
        vals = line.replace("\n", "").split(',')
        if cols is None:
            cols = vals
        else:
            data.append([float(x) for x in vals])
            d0 = pd.DataFrame(data, columns=cols)
```

```python
d0
```

```python
d0.dtypes
```

# method1 - numpy loadtxt

```python
d1 = np.loadtxt(filename, skiprows=1, delimiter=",")
```

```python
d1
```

```python
d1.dtype  # this is an array not a dataframe
```

# method2 - numpy genfromtxt

```python
#dtype=None has function infer datatype, so it treats the last column as integer
d2 = np.genfromtxt(filename, delimiter=",", names=True, dtype=None)
```

```python
d2
```

```python
d2.dtype
```
